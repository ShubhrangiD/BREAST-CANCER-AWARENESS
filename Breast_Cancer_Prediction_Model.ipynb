```python
import pandas as pd
import numpy as np
```

```python
breast = pd.read_csv('breast_data.csv')
breast.head()
```

```python
breast['diagnosis'].value_counts() #M = malignant or B = benign.
```

```python
breast.shape
```

```python
breast.isnull().sum()
```

```python
breast.duplicated().sum()
```

```python
breast.info()
```

```python
breast.drop('Unnamed: 32',axis=1,inplace = True)
```

```python
breast.describe() #statistic for numeric feature
```

Encoding the Target Variable

```python
#from sklearn.preprocessing import LabelEncoder
#Le = LabelEncoder()
breast['diagnosis']=breast['diagnosis'].map({"M":1,"B":0})
breast['diagnosis'].value_counts()
```

```python
breast.corr()
```

splitting ata into training and testing sets

```python
x = breast.drop('diagnosis',axis=1)
y=breast['diagnosis']
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python

sc.fit(x_train)
```

```python
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

Training Model

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
y_pred = lg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
